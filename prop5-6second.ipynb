```sage
n = 9
subsetsR = Subsets(list(range(1, n+1)))
power_set = [[0] + list(x) for x in subsetsR if x]

sum_result = [0] * (n - 1)
for R in power_set:
    prod = [1]
    for i in range(1, len(R)):
        stirnums = [0] + [stirling_number2(n - R[i-1], n - R[i])] * (R[i] - R[i-1] - 1)
        prodtemp = prod.copy()
        prod = convolution(prodtemp, stirnums)

    prod = prod + [0] * (n - len(prod) - 1)
    sum_result_temp = sum_result.copy()
    sum_result = [x + y for x, y in zip(sum_result_temp, prod)]
    print("R:", R, "prod:", prod, "sumResult:", sum_result)

sum_result
```
